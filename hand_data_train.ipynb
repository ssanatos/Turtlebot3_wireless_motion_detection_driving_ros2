```python
import pandas as pd
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

```python
# 정확도
def get_clf_eval(y_test, y_pred):
    accuracy = accuracy_score(y_test, y_pred)
    print('\n정확도: {:.4f}'.format(accuracy))
```

```python
# 훈련 데이터
incr1 = pd.read_csv('./hand_incr1.csv')
decr1 = pd.read_csv('./hand_decr1.csv')
back1 = pd.read_csv('./hand_back1.csv')
stab1 = pd.read_csv('./hand_stab1.csv')
incr2 = pd.read_csv('./hand_incr2.csv')
decr2 = pd.read_csv('./hand_decr2.csv')
back2 = pd.read_csv('./hand_back2.csv')
stab2 = pd.read_csv('./hand_stab2.csv')
dataset = pd.concat([incr1,decr1,back1,stab1,incr2,decr2,back2,stab2], axis=0, ignore_index = True)
dataset = dataset.dropna()
dataset
```

```python
X_features = dataset.drop(['order'],axis=1)
y_label = dataset['order']
X_train, X_test, y_train, y_test = train_test_split(X_features, y_label, test_size=0.2, random_state=156)
xgb_wrapper = XGBClassifier(n_estimators = 400, learning_rate = 0.1, max_depth = 4)
xgb_wrapper.fit(X_train, y_train)
yhat = xgb_wrapper.predict(X_test)
print(get_clf_eval(y_test, yhat))
```
